LEVEL 1:

```python
# Global stacks to maintain data
stack = []
min_stack = []
max_stack = []

def push(x):
    """Pushes element x onto the stack."""
    stack.append(x)

    # Maintain the min stack
    if not min_stack or x <= min_stack[-1]:
        min_stack.append(x)
    else:
        min_stack.append(min_stack[-1])

    # Maintain the max stack
    if not max_stack or x >= max_stack[-1]:
        max_stack.append(x)
    else:
        max_stack.append(max_stack[-1])

def pop():
    """Removes the top element of the stack."""
    if stack:
        stack.pop()
        min_stack.pop()
        max_stack.pop()

def top():
    """Returns the top element without removing it."""
    return stack[-1] if stack else "Stack is empty"

def getMin():
    """Returns the smallest element in the stack."""
    return min_stack[-1] if min_stack else "Stack is empty"

def getMax():
    """Returns the largest element in the stack."""
    return max_stack[-1] if max_stack else "Stack is empty"

# Handling user input
while True:
    print("\nChoose an operation:")
    print("1. Push an element")
    print("2. Pop an element")
    print("3. Get top element")
    print("4. Get minimum element")
    print("5. Get maximum element")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == "1":
        value = int(input("Enter the element to push: "))
        push(value)
        print(f"Pushed {value} onto the stack.")
    elif choice == "2":
        if stack:
            pop()
            print("Popped the top element.")
        else:
            print("Stack is empty. Nothing to pop.")
    elif choice == "3":
        print("Top element:", top())
    elif choice == "4":
        print("Minimum element:", getMin())
    elif choice == "5":
        print("Maximum element:", getMax())
    elif choice == "6":
        print("Exiting program...")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 6.")
```

```python
# List to store non-overlapping sorted intervals
intervals = []

def binary_search(target):
    """Finds the correct index to insert an interval using binary search."""
    low, high = 0, len(intervals) - 1
    while low <= high:
        mid = (low + high) // 2
        if intervals[mid][0] < target:
            low = mid + 1
        else:
            high = mid - 1
    return low  # Correct position to insert the interval

def addInterval(start, end):
    """Adds a new interval [start, end] and merges overlapping intervals efficiently."""
    index = binary_search(start)  # Find insertion point

    # Merge overlapping intervals
    if index > 0 and intervals[index - 1][1] >= start:
        index -= 1
        start = min(start, intervals[index][0])
        end = max(end, intervals[index][1])

    # Merge subsequent overlapping intervals
    while index < len(intervals) and intervals[index][0] <= end:
        end = max(end, intervals[index][1])
        intervals.pop(index)

    # Insert the merged interval at the correct position
    intervals.insert(index, [start, end])

def getIntervals():
    """Returns the current list of non-overlapping, sorted intervals."""
    return intervals

# Handling user input
while True:
    print("\nChoose an operation:")
    print("1. Add an interval")
    print("2. Get all intervals")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")

    if choice == "1":
        start, end = map(int, input("Enter interval start and end (space-separated): ").split())
        addInterval(start, end)
        print("Interval added and merged if necessary.")
    elif choice == "2":
        print("Current non-overlapping intervals:", getIntervals())
    elif choice == "3":
        print("Exiting program...")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 3.")
```

```python
import time
import heapq

# Global dictionary to store key-value pairs with expiry timestamps
cache = {}
expiry_heap = []  # Min-heap to store expiry times for cleanup

def set(key, value, expiryTime):
    """Stores the key-value pair with an expiration timestamp."""
    global cache, expiry_heap
    expiry_timestamp = time.time() + expiryTime  # Calculate expiry time

    # Store or update the key in the cache
    cache[key] = (value, expiry_timestamp)

    # Push the expiry time into the heap
    heapq.heappush(expiry_heap, (expiry_timestamp, key))

    # Cleanup expired keys
    cleanup_expired_keys()

def get(key):
    """Retrieves the value if it exists and hasn't expired, else returns None."""
    cleanup_expired_keys()  # Remove expired keys before fetching

    if key in cache:
        value, expiry_timestamp = cache[key]
        if expiry_timestamp > time.time():
            return value  # Return value if not expired

    return None  # Return None if key does not exist or is expired

def cleanup_expired_keys():
    """Removes expired keys from the cache using the heap."""
    global cache, expiry_heap
    current_time = time.time()

    while expiry_heap and expiry_heap[0][0] <= current_time:
        _, expired_key = heapq.heappop(expiry_heap)  # Get the expired key
        if expired_key in cache:
            del cache[expired_key]  # Remove expired key from dictionary

# Handling user input
while True:
    print("\nChoose an operation:")
    print("1. Set a key-value pair with expiry")
    print("2. Get a value by key")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")

    if choice == "1":
        key = input("Enter key: ")
        value = input("Enter value: ")
        expiry_time = int(input("Enter expiry time in seconds: "))
        set(key, value, expiry_time)
        print(f"Stored key '{key}' with expiry in {expiry_time} seconds.")

    elif choice == "2":
        key = input("Enter key to retrieve: ")
        result = get(key)
        if result is not None:
            print(f"Value for '{key}': {result}")
        else:
            print(f"Key '{key}' not found or expired.")

    elif choice == "3":
        print("Exiting program...")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 3.")
```
